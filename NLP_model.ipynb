```python
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LSTM
from keras.layers import Embedding
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
rev=pd.read_csv('/content/drive/My Drive/review.csv')
lab=pd.read_csv('/content/drive/My Drive/label.csv')
```

```python
split_frac = 0.80
split_idx = int(len(rev)*split_frac)
train_x, remaining_x = rev[:split_idx], rev[split_idx:]
train_y, remaining_y = lab[:split_idx], lab[split_idx:]
test_idx = int(len(remaining_x)*0.5)
val_x, test_x = remaining_x[:test_idx], remaining_x[test_idx:]
val_y, test_y = remaining_y[:test_idx], remaining_y[test_idx:]
print("\t\t\tFeature Shapes:")
print("Train set: \t\t{}".format(train_x.shape), 
      "\nValidation set: \t{}".format(val_x.shape),
      "\nTest set: \t\t{}".format(test_x.shape))
```

```python
model = Sequential()
model.add(Embedding(10262, 128,input_length=20))
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(train_x, train_y,
          batch_size=30,
          epochs=15,
          validation_data=(val_x, val_y))
```

```python
score, acc = model.evaluate(test_x,test_y,batch_size=128)
print("Score : ",score)
print("Accuracy : ",acc)
```
